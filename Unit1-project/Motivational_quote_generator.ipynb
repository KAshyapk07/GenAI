
## 21.	Motivational Quote Generator
o	Goal: Generate unique motivational quotes based on keywords like "Success", "Grind", "Peace".  
o	Tech: gpt2 with high temperature for creativity.


```python

from transformers import pipeline
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2",
)
```

```python
def generate_quote(keyword):
    """
    Generate a motivational quote based on a keyword (GPT-2 optimized).
    """

    prompt = f''' short and simple Motivational quote about {keyword}:

"{keyword.capitalize()} '''

    output = generator(
        prompt,
        max_new_tokens=40,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.8,
        top_k=40,
        top_p=0.9,
        repetition_penalty=1.2,
        pad_token_id=50256
    )

    generated_text = output[0]["generated_text"]

    quote = generated_text.replace(prompt, "").strip()

    for stop_char in ['"', '\n']:
        if stop_char in quote:
            quote = quote.split(stop_char)[0]

    quote = quote.strip()

    if not quote.endswith(('.', '!', '?')):
        quote += '.'

    return f'"{keyword.capitalize()} {quote}'
```

```python

print("   AI MOTIVATIONAL QUOTE GENERATOR")
print("="*50)
print("\nType a keyword (e.g., Success, Grind, Peace)")
print("or 'exit' to quit.\n")

while True:
    user_input = input("Enter keyword: ").strip()
    
    if user_input.lower() in ['exit', 'quit', 'q']:
        print("\nThank you for using the quote generator!")
        break
        
    if not user_input:
        print("Please enter a valid keyword.\n")
        continue
        
    print(f"\n Generating quote for '{user_input}'...\n")
    
    try:
        # Use the improved version that generates multiple options
        quote = generate_quote(user_input)
        print(f" {quote}\n")
        
    except Exception as e:
        print(f" Error generating quote: {e}\n")
```
